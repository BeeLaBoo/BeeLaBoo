```python
## find null value in any column then replace it with the mean of that column

car['Cylinders'].fillna(car['Cylinders'].mean(), inplace=True)
```

```python
car.isnull().sum()
```

```python
car.head(5)
```

```python
car['Make'].fillna(car['Make'], inplace=True);  ## not int : cannot calculate MEAN
car['Model'].fillna(car['Model'], inplace=True);
car['Type'].fillna(car['Type'], inplace=True);
car['DriveTrain'].fillna(car['DriveTrain'], inplace=True);
car['MSRP'].fillna(car['MSRP'], inplace=True);
car['Invoice'].fillna(car['Invoice'], inplace=True);
car['EngineSize'].fillna(car['EngineSize'], inplace=True);
car['Horsepower'].fillna(car['Horsepower'], inplace=True);
car['MPG_City'].fillna(car['MPG_City'], inplace=True);
car['MPG_Highway'].fillna(car['MPG_Highway'], inplace=True);
car['Weight'].fillna(car['Weight'], inplace=True);
car['Wheelbase'].fillna(car['Wheelbase'], inplace=True);
car['Length'].fillna(car['Length'], inplace=True);
```

```python
car.isnull().sum()
```
