```python
import requests
import pandas as pd
import lxml
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://www.carpages.ca/used-cars/search/?transmissiontype_id=1&province_code=on&sale_type=used&keyword=hybrid')
```

```python
soup = BeautifulSoup(page.text,'lxml')
soup
```

list_link

while True :
    next_page = (soup.find_all('a','href'))
    url = next_page
    page = requests.get(url)
    soup = BeautifulSoup(page.text,'lxml')

```python
# find all posting
postings = soup.find_all('div',class_='media soft push-none rule')
len(postings)

# 49 posting
```

```python
# dataframe ({[],[],[]}) , **'','','' : they are coloumn names
df =pd.DataFrame({'Name':[''],'Color':[''],
                 'Price':[''],'Link':['']}) 
##: name,color...they are list[]already ,no need []

#create for loop to go through each posting__to get name,color,price from each post

for post in postings:
    name = post.find('h4',class_='hN').text.strip()
     # 2'grey l-column l-column--small-6 l-column--medium-4' but I want index1
    color = post.find_all('div',class_='grey l-column l-column--small-6 l-column--medium-4')[1].text.strip()
    price = post.find('strong',class_='delta').text.strip()
    link = post.find('a',class_='media__img media__img--thumb').get('href')
    link_full = ("https://www.carpages.ca"+link)
    
    df = df.append({'Name':name,'Color':color,
                 'Price':price,'Link':link_full},ignore_index=True)
    
    ###### dont forget : ignore_index=True ######
    
##then put in dataframe , on the top
```
