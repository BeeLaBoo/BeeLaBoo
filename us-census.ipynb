## ISIN FUNCTION

```python
##data.head()show record of Michigan , Ohio , FL

data[data['State'].isin(['Michigan','Ohio','Florida'])]
```

```python
## total population by State

data.groupby('State').TotalPop.sum().sort_values(ascending = False)
```

```python
## group by state and work at home , self employed

data.groupby('State')['WorkAtHome','SelfEmployed'].sum().sort_values(by =['WorkAtHome','SelfEmployed'],ascending = False)
```
