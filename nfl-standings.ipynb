```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import lxml
import re
  
```

```python
#Imports the HTML into python
url = 'https://www.nfl.com/standings/league/2021/REG'
page = requests.get(url)
page

```

```python

soup = BeautifulSoup(page.text, 'lxml')
```

```python
soup
```

```python
table = soup.find('table', {'summary':'Standings - Detailed View'})
table
```

```python
# previously , I got whole table , then find header
table.find_all('th')

# th tags are all in the first row - header in html script
```

```python
headers =[]
for i in table.find_all('th'):
    title = i.text.strip()
    headers.append(title)
```

```python
headers
```

```python
df = pd.DataFrame(columns = headers)
```

```python
df
```

#grab data to put in each headers

# want each data in text format : data inside all in td tag
# and will create for loop : get list of all data row by row



```python
table.find_all('tr')
for row in table.find_all('tr')[1:-1]: 
    
            
    #don'want first row (column name), looping row by row
    data = row.find_all('td')
    
    #data in each row, get all td
    data_row = [td.text.strip() for td in data]    # list of td
    
   
```

```python
data_row
   
```

```python
import pandas as pd

df = pd.DataFrame(columns=headers)

df.loc[len(df)] = data_row

```

```python
df.to_csv("NFL_stand.csv")
```
