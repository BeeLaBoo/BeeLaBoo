```python
from bs4 import BeautifulSoup
import lxml
import requests

url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers'

requests.get(url)
page = requests.get(url)
# asking this host : Can I grab this html
page.text

soup = BeautifulSoup(page.text,'lxml')
#grabbing html overthere as a text and put in python xml format
soup.status_code

print(soup)
#now I get html in Python 
# how I can make it look prettier
soup.header
soup.div

#navigable string
tag = soup.header.p
tag.string
print(tag.string)

tag = soup.header.a
print(tag)

# get attribute names 
tag.attrs
print(tag.attrs)
###--dict of distribute names : {'data-toggle': 'collapse-side', 'data-target':
    #--'.side-collapse', 'data-target-2': '.side-collapse-container'}--###

tag['data-toggle']
print(tag['data-toggle']) #collapse-side

#add new attribute
tag['new_attr'] = "this is new attribute"
tag.attrs
print(tag.attrs)
#{'data-toggle': 'collapse-side', 'data-target': '.side-collapse', 'data-target-2': '.side-collapse-container', 'new_attr': 'this is new attribute'}
```
